### Module Level dunder Names

A module level dunder like (__all__, __author__, __version__) should be placed at the module main docstring and should be before all the import statements.

```python
from __future__ import print
__all__ = ['searching', 'sorting']
__version__ = '0.0.1'
__author__ = 'Chitrank Dixit'  
import os 
import sys
```
